```python
import numpy as np
import matplotlib.pyplot as plt
from surprise.model_selection import train_test_split
from surprise import Dataset, Reader, SVD, accuracy

n_factors = 20
n_epochs = 20

# Initialize lists to store training and test RMSE values
train_rmse_values = []
test_rmse_values = []

# Train the SVD model and calculate RMSE at each epoch
for epoch in range(1, n_epochs + 1):
    svd = SVD(n_factors=n_factors, n_epochs=epoch, reg_all=0.01, random_state=42)
    svd.fit(trainset)
    
    train_predictions = svd.test(trainset.build_testset())
    train_rmse = accuracy.rmse(train_predictions, verbose=False)
    train_rmse_values.append(train_rmse)
    
    test_predictions = svd.test(testset)
    test_rmse = accuracy.rmse(test_predictions, verbose=False)
    test_rmse_values.append(test_rmse)

# Plot the learning curves
epochs = np.arange(1, n_epochs + 1)
plt.plot(epochs, train_rmse_values, marker='o', label='Training RMSE')
plt.plot(epochs, test_rmse_values, marker='s', label='Test RMSE')
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.title('Learning Curves for SVD Model')
plt.grid()
plt.legend()
plt.show()
```
